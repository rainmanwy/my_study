```python
"""策略模式(Strategy Pattern):
定义一系列算法,把它们一一封装起来,并且使它们可以相互替换。本模式使得算法可以独立于使用它的客户而变化。

下面的例子，是不同条件下的折扣，通过不同实现类，来实现具体策略
"""
from abc import ABC, abstractmethod
from collections import namedtuple

Customer = namedtuple('Customer', 'name fidelity')

class LineItem:
    def __init__(self, product, quantity, price):
        self.product = product
        self.quantity = quantity
        self.price = price
        
    def total(self):
        return self.price * self.quantity

class Order:
    # 上下文
    def __init__(self, customer, cart, promotion=None):
        self.customer = customer
        self.cart = list(cart)
        self.promotion = promotion
    
    def total(self):
        if not hasattr(self, '__total'):
            self.__total = sum(item.total() for item in self.cart)
        return self.__total

    def due(self):
        if self.promotion is None:
            discount = 0
        else:
            discount = self.promotion.discount(self)
        return self.total() - discount

    def __repr__(self):
        fmt = '<Order total: {:.2f} due: {:.2f}>'
        return fmt.format(self.total(), self.due())
    
class Promotion(ABC) : # 策略:抽象基类
        
        @abstractmethod
        def discount(self, order):
            """返回折扣金额(正值)"""
            pass

class FidelityPromo(Promotion): # 第一个具体策略
    """为积分为1000或以上的顾客提供5%折扣"""

    def discount(self, order):
        return order.total() * .05 if order.customer.fidelity >= 1000 else 0

class BulkItemPromo(Promotion): # 第二个具体策略
    """单个商品为20个或以上时提供10%折扣"""
    def discount(self, order):
        discount = 0
        for item in order.cart:
            if item.quantity >= 20:
                discount += item.total() * .1
        return discount

class LargeOrderPromo(Promotion): # 第三个具体策略
    """订单中的不同商品达到10个或以上时提供7%折扣"""
    def discount(self, order):
        distinct_items = {item.product for item in order.cart}
        if len(distinct_items) >= 10:
            return order.total() * .07
        return 0

joe = Customer('John Joe', 0)
ann = Customer('Ann', 1100)
cart = [
    LineItem('banana', 4, 0.5),
    LineItem('apple', 10, 1.5)
]
print(Order(joe, cart, FidelityPromo()))
```

```python
"""
functools.lru_cache: 现了备忘(memoization)功能。这是
一项优化技术,它把耗时的函数的结果保存起来,避免传入相同的参数时重复计算。
"""

import functools
import time

@functools.lru_cache()
def fibo(n):
    if n < 2:
        return n
    return fibo(n-2) + fibo(n-1)

def fibo_nocache(n):
    if n < 2:
        return n
    return fibo_nocache(n-2) + fibo_nocache(n-1)

begin = time.time()
fibo(35)
end = time.time()
print(end-begin)
begin = time.time()
fibo_nocache(35)
end = time.time()
print(end-begin)
```
